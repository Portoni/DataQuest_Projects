### **Detecting Outliers**

Outliers are not necesseraly worng data, they are just values with a big discrepancy from the rest, that reduces the accuracy of our analyses. So it is very import that we clean the data and remove outliers so we can have a more precise analyses for this project.

Looking at the histogram distribution, we can clearly see the presence of outliers in many columns. The columns we will analyse is `price`, `minimum_nights` and `calculated_host_listings_count`.

Thi values don't follow a distribution and distort the entire graphical representation. There is two quick methods to detect outliers:

- Statistical resume using `describe()`.
- Plot `boxplots` for the variable.

```
df[['price', 'minimum_nights', 'calculated_host_listings_count']].describe()
```

Investigating the statistical information, we can confirm the existence of outliers on this three columns.
- The variable `price` has 75% of it's values under 193, but your maximum value is 15000.
- The variable `minimum_nights` has 75% of it's values under 3, but your maximum value is 1125.
- The variable `calculated_host_listings_count` has 75% of it's values under 4, but your maximum value is 151.

Now, let's see how this data is presented in a boxplot.

#### Boxplot for minimum_nights

```
df.minimum_nights.plot(kind='box', vert=False, figsize=(16,2));
plt.show()
```

To set the range of useful data in the minimum nights column, we can use common sense, 30 days seems a reasonable number, let's see how many values are bigger than 30 and the percentage it represents from our data.

```
# amount of values over 30 days for minimum_nights
print('Number of minimum nights higher than 30 days: {}'.format(df.minimum_nights[df.minimum_nights > 30].count()))
print('It represents {:.4f}% of our data in this column'.format((df.minimum_nights[df.minimum_nights > 30].count() / df.shape[0]) * 100))
```

#### Boxplot for price

```
df.price.plot(kind='box', vert=False, figsize=(16,2));
```

Looking at the boxplot above, we can clearly see there are outliers in the `price` column, we will not use any statistical method to determine the range of values considered outliers because it is not in this project scope. The number used to determine the outliers will be determined by common sense.

```
# amount of values in the outlier range
print('Number of prices higher than 1000: {}'.format(df.price[df.price > 1000].count()))
print('It represents {:.4f}% of our data in this column'.format((df.price[df.price > 1000].count() / len(df.price) * 100)))
```

#### Removing the outliers

Now that we identify the outliers present in `minimum_nights` and `price` columns, let's clean the *DataFrame* and plot another histogram.

Let's remove the columns `license` and `neighbourhood_group`, they don't have any data.

```
df_clean = df.copy()

df_clean = df_clean.drop(df_clean.price[df_clean.price > 1000].index)
df_clean = df_clean.drop(df_clean.minimum_nights[df_clean.minimum_nights > 30].index)

# remove `license` and `neighbourhood_group`
df_clean = df_clean.drop('license', axis=1)
df_clean = df_clean.drop('neighbourhood_group', axis=1)

# Ploting a new histogram with the variables
df_clean.hist(bins=15, figsize=(15,10));
```

### **Extracting the correlation between variables**

Checking for correlation between variables is a good way of extracting information that can be used for generating value for a company.

This relation can be measured, and it is a function of the correlation coefficient to establish its intensity.

```
correlation_matrix = df_clean[['price', 'minimum_nights', 'number_of_reviews',
'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].corr()

display(correlation_matrix)
```

```
# heatmap correlation
sns.heatmap(correlation_matrix, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

From the heatmap we can see that the only strong correlation are between the variables `reviews_per_month` and `number_of_reviews`, the other variables don't have strong dependency.

### **Analyzing property types**

The column `room_type` indicates the kind of property that is being hosted on the Airbnb app. There are many options, like sharing rooms or get one room for yourself, even entire homes or apartments.

Let's see the percentage of each room type in the dataset.

```
df_clean.room_type.value_counts()
```

```
df_clean.room_type.value_counts() / df_clean.shape[0]
```

### **Analyzing price per Neighbourhood and room type**



It is important to explore the prices in relation to the location and room type that is being hosted, so the person looking to stay can choose the best options acording to his own criterias.

```
df_clean.price.mean()
```

```
# prices per Neighbourhood
df_clean.groupby('neighbourhood')[['price']].mean().sort_values(['price'], ascending=False).head(10)
```

Most of the neighbourhoods above have the prices higher then the mean of our dataset overall.

```
# prices per room type
df_clean.groupby('room_type')[['price']].mean().sort_values(['price'], ascending=False)
```

Shared rooms are the best choice for people looking to spend the least amount of cash. Peoplo that want to have more privacy and don't bother spending more money, they can stay in hotel rooms.

A good data visualization is to plot all the propertys in function of their respective latitude and longitude.

```
df_clean.plot(kind='scatter', x='longitude', y='latitude', alpha=0.4,
              c=df_clean['price'], s=8, cmap=plt.get_cmap('jet'), figsize=(12,8));
```
